### Add Winner Column

```python
import pandas as pd

cols = ["title", "matchup", "url", "emcee1", "emcee2"]

# Convert list URLs to a single string so rows are hashable
df_review = df_battles.loc[:, cols].copy()
df_review["url"] = df_review["url"].apply(
    lambda x: " | ".join(x) if isinstance(x, list) else x
)

df_review = (
    df_review
    .drop_duplicates()
    .sort_values("matchup", na_position="last")
    .reset_index(drop=True)
)

df_review["winner"] = ""
df_review["judging"] = ""

out_path = "../data/raw/battle_winners_review.xlsx"
df_review.to_excel(out_path, index=False)

print(f"Wrote {len(df_review)} rows to {out_path}")
```
