```python
from fliptop import RAW_DATA_DIR
from fliptop.data_cleaning import (
    load_youtube_uploads,
    clean_titles,
    parse_upload_date,
    add_duration_columns,
    convert_video_metrics_to_numeric,
    copy_yt_title,
    strip_pt_suffix_from_title,
    filter_titles_with_vs,
    drop_non_battles,
    keep_1v1,
    add_matchup_and_split,
    apply_emcee_rename,
    add_matchup_clean,
)
from fliptop.rename_map import RENAME_MAP
```

```python
df_yt = load_youtube_uploads(RAW_DATA_DIR / "youtube_videos.json")
```

```python
df_1v1 = (
    df_yt
    .pipe(clean_titles)
    .pipe(parse_upload_date)
    .pipe(add_duration_columns)
    .pipe(convert_video_metrics_to_numeric)
    .pipe(copy_yt_title)
    .pipe(strip_pt_suffix_from_title)
    .pipe(filter_titles_with_vs)
    .pipe(drop_non_battles)
    .pipe(keep_1v1)
    .pipe(add_matchup_and_split)
    .pipe(apply_emcee_rename, rename_map=RENAME_MAP)
    .pipe(add_matchup_clean)
)
```

```python
df_1v1[["id", "title", "matchup", "emcee1", "emcee2", "matchup_clean"]].tail(20)
```
